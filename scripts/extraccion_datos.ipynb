```python
import pandas as pd
import numpy as np
import random

ciudades = ['Quito', 'Guayaquil', 'Cuenca', 'Ambato', 'Manta']
niveles_educacion = ['Primaria', 'Secundaria', 'Tecnologico', 'Universitario']
estados_laborales = ['Empleado', 'Desempleado', 'Subempleado']
sectores = ['Comercio', 'Servicios', 'Manufactura', 'Agricultura', 'Construccion', 'Tecnologia']

registros = []

for i in range(1, 5001):
    ciudad = random.choice(ciudades)
    edad = random.randint(18, 65)
    educacion = random.choice(niveles_educacion)
    estado = random.choice(estados_laborales)
    
    if estado == 'Empleado':
        ingreso = round(random.uniform(460, 2500), 2)
        meses_desempleado = 0
        sector = random.choice(sectores)
    elif estado == 'Subempleado':
        ingreso = round(random.uniform(100, 459), 2)
        meses_desempleado = 0
        sector = random.choice(sectores)
    else:
        ingreso = 0.0
        meses_desempleado = random.randint(1, 36)
        sector = 'Ninguno'
        
    registros.append([i, ciudad, edad, educacion, estado, ingreso, sector, meses_desempleado])

df_simulado = pd.DataFrame(registros, columns=[
    'ID_Persona', 'Ciudad', 'Edad', 'Nivel_Educativo', 
    'Estado_Laboral', 'Ingreso_Mensual', 'Sector_Economico', 'Meses_Desempleado'
])

ruta_guardado = r"C:\Users\USUARIO\Documents\GitHub\Proyecto_Analisis_Datos_2026\data\raw\datos_simulados_economia_desempleo.csv"
df_simulado.to_csv(ruta_guardado, index=False, encoding='latin1', sep=';')

df_simulado.head()
```

```python
import pandas as pd
import numpy as np

# Definimos el número exacto de registros
n_registros = 924958

# 1. Generamos datos base
data = {
    'id_registro': range(1, n_registros + 1),
    'anio': np.random.choice([2024, 2025, 2026], size=n_registros),
    'provincia': np.random.choice(['Pichincha', 'Guayas', 'Azuay', 'Manabi', 'Loja'], size=n_registros),
    'pea': np.random.randint(10000, 500000, size=n_registros),
    'salario_promedio': np.random.uniform(400, 2000, size=n_registros).round(2)
}

df = pd.DataFrame(data)

# 2. Introducimos errores intencionales para la limpieza
# --- Nulos ---
df.loc[df.sample(frac=0.02).index, 'pea'] = np.nan

# --- Duplicados ---
duplicados = df.sample(n=1000)
df = pd.concat([df, duplicados], ignore_index=True)

# --- Outliers (Errores de captura: salarios negativos) ---
df.loc[df.sample(frac=0.01).index, 'salario_promedio'] = -500.0

# --- Formatos incorrectos (Texto en columnas numéricas) ---
# Forzamos que algunas filas de 'pea' sean texto
df.loc[df.sample(frac=0.01).index, 'pea'] = "Error_Carga"

# Exportamos a CSV
nombre_archivo = 'dataset_economia_sucio.csv'
df.to_csv(nombre_archivo, index=False)

print(f"¡Listo! Archivo '{nombre_archivo}' creado con {len(df)} registros.")
```

```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
import urllib.parse

# 1. GENERACIÓN DEL DATASET SUCIO (tal cual lo tenemos)
n_registros = 924958
data = {
    'id_registro': range(1, n_registros + 1),
    'anio': np.random.choice([2024, 2025, 2026], size=n_registros),
    'provincia': np.random.choice(['Pichincha', 'Guayas', 'Azuay', 'Manabi', 'Loja'], size=n_registros),
    'pea': np.random.randint(10000, 500000, size=n_registros),
    'salario_promedio': np.random.uniform(400, 2000, size=n_registros).round(2)
}
df = pd.DataFrame(data)

# Introducimos los errores
df.loc[df.sample(frac=0.02).index, 'pea'] = np.nan
duplicados = df.sample(n=1000)
df = pd.concat([df, duplicados], ignore_index=True)
df.loc[df.sample(frac=0.01).index, 'salario_promedio'] = -500.0
df.loc[df.sample(frac=0.01).index, 'pea'] = "Error_Carga"

print("Dataset sucio generado en la variable 'df'.")

# 2. CONEXIÓN A LA BASE DE DATOS
usuario = 'root'
contrasena = '123456'
servidor = 'localhost'
puerto = '3306'
base_datos = 'proyectoeconomia'

contrasena_codificada = urllib.parse.quote_plus(contrasena)
cadena_conexion = f"mysql+pymysql://{usuario}:{contrasena_codificada}@{servidor}:{puerto}/{base_datos}"
motor_mysql = create_engine(cadena_conexion)

# 3. EXPORTACIÓN DEL DATAFRAME 'df' (EL SUCIO)
# Usamos 'df' (sin el _limpio) porque ese es el que tiene los datos con errores
df.to_sql('dataset_economia_sucio', con=motor_mysql, if_exists='replace', index=False, chunksize=5000)

print("¡Éxito! El dataset 'dataset_economia_sucio' ya está en la base 'proyectoeconomia'.")
```
