```python
import json
import random

# Definimos los datos base basados en tu contexto
periodos = ["1. Trimestre", "2. Trimestre", "3. Trimestre", "4. Trimestre"]
provincias = ["IMBABURA", "PICHINCHA", "AZUAY", "GUAYAS", "LOJA"]
servicios = ["Asistencia Técnica", "Servicio Limpieza", "Obtención Producto", "Mejora Flujo Proceso"]
sectores = ["Textil", "Agropecuario", "Servicios", "Comercio"]

def generar_datos(n):
    registros = []
    for i in range(n):
        registro = {
            "RowID": f"Row{i}",
            "Periodo": random.choice(periodos),
            "Mes": random.choice(["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio"]),
            "Provincia": random.choice(provincias),
            "Sector_Productivo": random.choice(sectores),
            "Servicio_DPC": random.choice(servicios),
            "No_Beneficiarios": random.randint(10, 100)
        }
        registros.append(registro)
    return registros

# Generar 1000 registros
datos = generar_datos(1000)

# Guardar en archivo JSON
with open('proyectos_ieps.json', 'w', encoding='utf-8') as f:
    json.dump(datos, f, ensure_ascii=False, indent=4)

print("¡Archivo 'proyectos_ieps.json' generado con éxito con 1000 registros!")
```
